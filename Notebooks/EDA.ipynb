```python
import numpy as np

X = np.load('/content/X.npy')
y = np.load('/content/y.npy')

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
print(f"Data type of X: {X.dtype}")
print(f"Data type of y: {y.dtype}")

print("\nFirst 5 rows of X:")
print(X[:5])

print("\nFirst 5 elements of y:")
print(y[:5])
```

```python
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)

print(f"Mean of each feature in X (first 5 values):\n{X_mean[:5]}")
print(f"\nStandard deviation of each feature in X (first 5 values):\n{X_std[:5]}")
print(f"\nMinimum of each feature in X (first 5 values):\n{X_min[:5]}")
print(f"\nMaximum of each feature in X (first 5 values):\n{X_max[:5]}")
```

```python
missing_X = np.sum(np.isnan(X))
missing_y = np.sum(np.isnan(y))

print(f"Total missing values in X: {missing_X}")
print(f"Total missing values in y: {missing_y}")

if missing_X > 0:
    print("Missing values are present in X.")
else:
    print("No missing values found in X.")

if missing_y > 0:
    print("Missing values are present in y.")
else:
    print("No missing values found in y.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the number of features to visualize
num_features_to_plot = 10

# Create subplots for better visualization
fig, axes = plt.subplots(nrows=num_features_to_plot, ncols=1, figsize=(10, 5 * num_features_to_plot))

for i in range(num_features_to_plot):
    sns.histplot(X[:, i], ax=axes[i], kde=True)
    axes[i].set_title(f"Distribution of Feature X[{i}]")
    axes[i].set_xlabel("Feature Value")
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
